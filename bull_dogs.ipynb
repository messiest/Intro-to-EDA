```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
plt.rcParams['font.size'] = 20
```

```python
dogs = pd.read_csv('Dogs of NYC - WNYC.csv')
```

```python
dogs.info()
```

```python
bull_dogs = dogs[dogs['breed'].str.contains('Bull Dog')]
bull_dogs['breed'].value_counts()
```

```python
sns.countplot(x="breed", data=bull_dogs);
plt.xlabel("Breed")
plt.ylabel("Count")
plt.title("Share of Bull Dogs in New York")
plt.show();
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="borough", hue="breed", data=bull_dogs);
plt.ylabel("Count")
plt.xlabel("")
plt.title("Share of Bull Dogs Across New York")
plt.legend(['English', 'French']).set_title('')
plt.show();
```

```python
pibbles = dogs[dogs['breed'].str.contains("Pit Bull") | dogs['breed'].isin(['American Pit Bull Terrier', 'Staffordshire Terrier', 'American Staffordshire Terrier', 'American Bully', 'Staffordshire Bull Terrier'])]
pibbles['breed'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="borough", hue="breed", data=pibbles);
plt.ylabel("Count")
plt.xlabel("")
plt.title("Share of Pit Bull 'Breeds' Across New York")
plt.legend().set_title('')
plt.show();
```
